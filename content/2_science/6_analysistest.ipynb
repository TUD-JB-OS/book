````python
fig1 = plt.figure()
fig1.plot(x, y, 'k.')
fig1.xlabel('x')
fig1.ylabel('y')

fig1.savefig('fig1.svg')
fig1_caption = 'some caption here' 
fig1_label = 'fig-no1'
fig1_width = '50%'
fig_name = 'fig1.svg'
plt.show()
````
