````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,10)
y = 2*x+3


fig1, ax = plt.subplots()     
ax.plot(x, y, 'k.')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig1.savefig('fig1.svg')
fig1_caption = 'some caption here' 
fig1_label = 'fig-no1'
plt.show()
````
