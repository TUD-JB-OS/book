````python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0,100,10)
y = 2*x+3

fig, ax = plt.subplots()      # fig = entire plot, ax = axes
ax.plot(x, y, 'k.')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.savefig('fig1.svg')
plt.show()
````
