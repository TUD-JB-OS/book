````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,10)
y = 2*x+3


fig, ax = plt.subplots()     
ax.plot(x, y, 'k.')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.savefig('fig1.svg')
caption1 = 'some caption here' 
plt.show()
````


````python
import numpy as np
import matplotlib.pyplot as plt
from iPython.display import Markdown

x = np.linspace(0,100,10)
y = 2*x+3


fig, ax = plt.subplots()     
ax.plot(x, y, 'k.')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.savefig('fig1.svg')
caption1 = 'some caption here' 

display(
    Markdown(
:::{figure} fig.show()
:label: fig-1

{eval}`caption1`
:::
    )
)

````
